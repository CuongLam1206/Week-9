```python
# câu 8:  xếp chồng 2 mảng theo chiều dọc
import numpy as np

# Tạo mảng hai chiều (2 hàng 5 cột) từ 0 đến 9
arr1 = np.arange(10).reshape(2, -1)
print('arr1: \n', arr1)

# Tạo mảng hai chiều (2 hàng 5 cột) chỉ có giá trị 1
arr2 = np.repeat(1, 10).reshape(2, -1)
print('arr2: \n', arr2)

# Cách 1: Dùng np.vstack() để xếp chồng theo chiều dọc
c = np.vstack((arr1, arr2))
print("Cách 1: \n", c)

# Cách 2: Dùng np.concatenate() với axis=0
d = np.concatenate((arr1, arr2), axis=0)
print("Cách 2: \n", d)

# Cách 3: Dùng np.row_stack() để xếp chồng theo chiều dọc
e = np.row_stack((arr1, arr2))
print("Cách 3: \n", e)
```

```python
# câu 9 : xếp chồng 2 mảng theo chiều ngang
import numpy as np

# Tạo mảng hai chiều (2 hàng 5 cột) từ 0 đến 9
arr1 = np.arange(10).reshape(2, -1)
print('arr1: \n', arr1)

# Tạo mảng hai chiều (2 hàng 5 cột) chỉ có giá trị 1
arr2 = np.repeat(1, 10).reshape(2, -1)
print('arr2: \n', arr2)

# Cách 1: Dùng np.hstack() để xếp chồng theo chiều ngang
c = np.hstack((arr1, arr2))
print("Cách 1: \n", c)

# Cách 2: Dùng np.concatenate() với axis=1
d = np.concatenate((arr1, arr2), axis=1)
print("Cách 2: \n", d)

# Cách 3: Dùng np.column_stack() để xếp chồng theo chiều ngang
e = np.column_stack((arr1, arr2))
print("Cách 3: \n", e)
```

```python
# câu 14 : Lấy tất cả các giá trị trong 1 phạm vi cho trước từ 1 mảng
import numpy as np

a = np.array([2, 6, 1, 9, 10, 3, 27])

# Tìm tất cả vị trí của các phần tử có giá trị trong phạm vi [5,10]

# Cách 1: Dùng np.where() để lấy chỉ số các phần tử trong phạm vi [5, 10]
index = np.where((a >= 5) & (a <= 10))
print("Cách 1: \n", a[index])

# Cách 2: Dùng np.nonzero() để lấy chỉ số các phần tử trong phạm vi [5, 10]
index = (a >= 5) & (a <= 10)
print("Cách 2: \n", a[index])

# Cách 3: Trực tiếp lấy kết quả các phần tử trong phạm vi [5, 10]
result = a[(a >= 5) & (a <= 10)]
print("Cách 3: \n", result)
```

```python
# Câu 21
import numpy as np

# Cho mảng a chứa các số ngẫu nhiên từ [1, 50)
a = np.random.randint(1, 50, size=10)  # Ví dụ mảng a có 10 phần tử ngẫu nhiên từ 1 đến 50
print("Mảng a:\n", a)

# Cách 1: Dùng hàm np.clip để thay thế giá trị lớn hơn 30 thành 30 và dưới 10 thành 10
result = np.clip(a, 10, 30)
print("Cách 1: \n", result)

# Cách 2: Dùng hàm np.where để thay thế giá trị lớn hơn 30 thành 30 và dưới 10 thành 10
result = np.where(a > 30, 30, np.where(a < 10, 10, a))
print("Cách 2: \n", result)
```

```python
# câu 22: Lấy vị trí của n giá trị đứng đầu từ một mảng numpy

import numpy as np

# Cho mảng a chứa các số ngẫu nhiên từ [1, 50)
a = np.random.randint(1, 50, size=10)
print("Mảng a:\n", a)

# Lấy vị trí của 5 giá trị lớn nhất trong một mảng a

# Cách 1: Dùng hàm np.argsort()
print("Cách 1:")
# Lấy vị trí của các phần tử lớn nhất
indices_sorted = np.argsort(a)[-5:]  # Lấy 5 chỉ số cuối cùng trong mảng đã sắp xếp
values_sorted = a[indices_sorted]    # Lấy giá trị tương ứng từ mảng a
print('Vị trí: ', indices_sorted)
print('Giá trị: ', values_sorted)

# Cách 2: Dùng hàm np.argpartition()
print("Cách 2:")
# Lấy vị trí của các phần tử lớn nhất bằng np.argpartition
indices_partition = np.argpartition(a, -5)[-5:]  # Lấy 5 chỉ số lớn nhất
values_partition = a[indices_partition]          # Lấy giá trị tương ứng từ mảng a
print('Vị trí: ', indices_partition)
print('Giá trị: ', values_partition)
```

```python
# câu 23: Chuyển đổi một mảng chứa nhiều mảng thành một mảng 1d phẳng
import numpy as np

# Cho array_of_arrays như sau
arr1 = np.arange(3)
arr2 = np.arange(3,7)
arr3 = np.arange(7,10)

array_of_arrays = np.array([arr1, arr2, arr3], dtype=object)
print('array_of_arrays: ', array_of_arrays)

# Cách 1: Dùng np.concatenate() để chuyển đổi thành mảng 1D phẳng
arr_2d = np.concatenate(array_of_arrays)
print("Cách 1: \n", arr_2d)

# Cách 2: Dùng np.flatten() hoặc np.ravel() để chuyển đổi thành mảng 1D phẳng
arr_2d = array_of_arrays.flatten()
print("Cách 2: \n", arr_2d)

# Cách 3: Dùng np.hstack() để chuyển đổi thành mảng 1D phẳng
arr_2d = np.hstack(array_of_arrays)
print("Cách 3: \n", arr_2d)

```

```python
# câu 24: Tìm giá trị lớn nhất trong mỗi hàng của mảng 2d
import numpy as np

# Cho mảng a
a = np.random.randint(1, 10, [5, 3])
print(a)

# Cách 1: Dùng np.amax để tìm giá trị lớn nhất trong mỗi hàng
result = np.amax(a, axis=1)
print("Cách 1: ", result)

# Cách 2: Dùng np.apply_along_axis và np.max để tìm giá trị lớn nhất trong mỗi hàng
result = np.apply_along_axis(np.max, axis=1, arr=a)
print("Cách 2: ", result)
```

```python
# câu 26: Tìm các vị trí có giá trị trùng lặp trong một mảng numpy

import numpy as np

# Cho mảng a
a = np.random.randint(0, 5, 10)
print(a)

# Tạo mảng out gồm tất cả giá trị là True
out = np.full(a.shape, True)

# Tìm tất cả các giá trị duy nhất và chỉ số lần đầu tiên xuất hiện
unique_a, return_index = np.unique(a, return_index=True)

# Đánh dấu các chỉ số lần đầu tiên xuất hiện là False
out[return_index] = False

# In ra kết quả
print(out)
```

```python
# câu 29: Tìm tất cả các cực đại cục bộ (hay còn gọi các đỉnh) trong một mảng 1d
import numpy as np

# Cho mảng a
a = np.array([1, 3, 7, 1, 2, 6, 0, 1])

# Tính sự thay đổi giữa các phần tử liên tiếp
doublediff = np.diff(np.sign(np.diff(a)))

# Tìm các cực đại (đỉnh) là các điểm có sự thay đổi từ dương sang âm
peak_locations = np.where(doublediff == -2)[0] + 1  # +1 để điều chỉnh vị trí vì np.diff giảm chiều dài mảng

print(peak_locations)
```
